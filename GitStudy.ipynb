3. COMMANDS

  3.1 CREATE

  * git init  = create a new repository
  * git clone = create a copy of the original repo on local machine
  * git fork  = create a copy of the original repo on GitHub account

  3.2 SYNC

  * git origin = [git remote add origin (repo link)] add a link of a remote repo to out local repo

    to check if linked:
    * git remote -v

  * git pull   = [git pull origin (master)] copy all the files from master branch of a remote repo to your local repo.
  * git push   = [gut push origin (master)] push local changes to central repo

  3.3 ADD

  * git add    = [git add (options) (ficheiro.txt)] add o ficheiro ao "staging area"

    -> . : um "ponto" aps o comando "add" aplica para todos os ficheiros alterados/criados

  * git commit = [git commit (optionals)] add ao local repo
    
    -> -m "somthing": comments

  3.4 PARALLEL DEVELOPMENT

  a) BRANCH
  #branch is the integral version of a repo, you can create a kind of copy branch, that works with snapshot system. When you need to make changes, but you dont wanna do at the principal branch, to crate others.
  #todas os commits feitos na masterbranch so tb copiados para as branchs alternativas. Master branch is used mostly just to production. the final aproved code.

  * git branch                              = check all the brunches
    -a (list remote and local)
  * git branch (nome)                       = create a branch
  * git checkout (nome)                     = migra pra branch escolhida
  * git push origin (namedabranch)          = envia a branch escolhida
  * git push origin --delete (namedabranch) = deleta do repo remoto
  * git branch -d (namedabranch)            = deleta do repo local
  * git branch -m oldbranch newbranch       = rename from local
  * git push origin :oldbranch newbranch    = rename from remoto

  b)MERGE

  #transfere as alteraes feitas na branch alternativa para a master. O merge  feito logado na branch master

  * git merge (nomedanewbranch) = transfere alteraes para a master
  
  c) REBASE

  #used when change made in one branche needs to be reflected in another branche

  * git rebase master

  d) FORK

  * git fork
   
    usado para copiar repositorio. Permite experiemtnar mudanas sem afetar o projeto original
